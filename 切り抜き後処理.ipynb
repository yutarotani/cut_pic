```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
id = 'man2.png'
img = cv2.imread(r'C:\Users\fminyu\Desktop\python\indexnet_matting\examples\images\man2.png')
img = img[...,::-1]
matte = cv2.imread(r'C:\Users\fminyu\Desktop\python\indexnet_matting\examples\mattes\man2.png')
h,w,_ = img.shape
bg = np.full_like(img,255) #white background
```

```python
img = img.astype(float)
bg = bg.astype(float)

matte = matte.astype(float)/255
img = cv2.multiply(img, matte)
bg = cv2.multiply(bg, 1.0 - matte)
outImage = cv2.add(img, bg)
outImage=(outImage/255)
Img = np.clip(outImage * 255, 0, 255).astype(np.uint8)
plt.imshow(Img)
plt.imsave('aa.jpg',Img)
```
